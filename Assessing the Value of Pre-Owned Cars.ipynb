```python
# importing modules
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import scipy as sp 
```

```python
df = pd.read_csv('data.csv')  
```

```python
df.head() 
```

```python

headers = ["symboling", "normalized-losses", "make",  
           "fuel-type", "aspiration","num-of-doors", 
           "body-style","drive-wheels", "engine-location", 
           "wheel-base","length", "width","height", "curb-weight", 
           "engine-type","num-of-cylinders", "engine-size",  
           "fuel-system","bore","stroke", "compression-ratio", 
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"] 
  
df.columns=headers 
df.head()
```

```python
# finding missing values
data = df     
data.isna().any() 
data.isnull().any() 
```

```python
# converting mpg to L / 100km 
data['city-mpg'] = 235 / df['city-mpg'] 
data.rename(columns = {'city_mpg': "city-L / 100km"}, inplace = True) 

print(data.columns) 
data.dtypes 
```

```python
data.price.unique()
```

```python
data = data[data.price != '?'] 
 
data.dtypes
```

```python
#converting 'price' column to numeric
data['price'] = pd.to_numeric(data['price'], errors='coerce')
```

```python
#droping rows with NaN values in the 'price' column
data = data.dropna(subset=['price'])
```

```python
# normalizaing amd creating bins for 'price' columns
data['length'] = data['length'] / data['length'].max()
data['width'] = data['width'] / data['width'].max()
data['height'] = data['height'] / data['height'].max()

bins = np.linspace(data['price'].min(), data['price'].max(), 4)
group_names = ['Low', 'Medium', 'High']
data['price-binned'] = pd.cut(data['price'], bins, labels=group_names, include_lowest=True)
```

```python
print(data['price-binned'])
```

```python
#plotting a histogram of the binned price data
plt.hist(data['price-binned'])
plt.xlabel('Price Bins')
plt.ylabel('Frequency')
plt.title('Histogram of Binned Prices')
plt.show()
```

```python
pd.get_dummies(data['fuel-type']).head() 

data.describe() 
```

```python
# plotting a box plot
# examples of box plot 
plt.boxplot(data['price']) 
```

```python
sns.boxplot(x ='drive-wheels', y ='price', data = data) 
 
plt.scatter(data['engine-size'], data['price']) 
plt.title('Scatterplot of Enginesize vs Price') 
plt.xlabel('Engine size') 
plt.ylabel('Price') 
plt.grid() 
plt.show() 
```

```python
test = data[['drive-wheels', 'body-style', 'price']] 
data_grp = test.groupby(['drive-wheels', 'body-style'], as_index = False).mean() 

data_grp 
```

```python
data_pivot = data_grp.pivot(index = 'drive-wheels', 
                            columns = 'body-style') 
data_pivot 
```

```python
#plotting a heatmap
plt.pcolor(data_pivot, cmap ='RdBu') 
plt.colorbar() 
plt.show() 
```

```python
data_annova = data[['make', 'price']] 
grouped_annova = data_annova.groupby(['make']) 
annova_results_l = sp.stats.f_oneway( 
                             grouped_annova.get_group('honda')['price'], 
                             grouped_annova.get_group('subaru')['price'] 
                                    ) 
print(annova_results_l) 

# here anova test return f test and p value, where f test represents variance and p value represents confidence degree
```
